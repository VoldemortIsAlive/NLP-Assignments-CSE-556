## Part A
### Question a

```python
#Assumption: sentence ends with punctuation . or ! or ?
regex_sentence = r'\b[^!\?\.]+[!\?\.]?'
regex_token = r'\b\S+\b'
```

```python
for class_label in [0,1]:
    count_sentences = 0
    count_tokens = 0
    num = 0
    for row in data:
        if row[0] != class_label:
            continue
        num += 1
        sentences = re.findall(regex_sentence, row[2])
        count_sentences += len(sentences)
        tokens = re.findall(regex_token, row[2])
        count_tokens += len(tokens)
#         print(tokens)
    avg_sentences = count_sentences/num
    avg_tokens = count_tokens/num
    print("For Class Label", class_label, ":")
    print("Average Sentences =", avg_sentences)
    print("Average Tokens =", avg_tokens)
```

### Question b
total number of words starting with consonants and vowels.

```python
regex_consonant = r'\b[bcdfghjklmnpqrstvwxyzBCDFGHJKLMNPQRSTVWXYZ]\S*\b'
regex_vowel = r'\b[aeiouAEIOU]\S*\b'
```

```python
for class_label in [0,1]:
    count_consonants = 0
    count_vowels = 0
    for row in data:
        if row[0] != class_label:
            continue
        consonants = re.findall(regex_consonant, row[2])
        count_consonants += len(consonants)
        vowels = re.findall(regex_vowel, row[2])
        count_vowels += len(vowels)
#         print(consonants)
    print("For Class Label", class_label, ":")
    print("Words starting with consonants =", count_consonants)
    print("Words starting with vowels =", count_vowels)
```

lowercase the text and report the number of unique tokens present before and
after lower casing.

```python
regex_token = r'\b\S+\b'
```

```python
#Before lowercasing:
print("Before Lowercasing")
for class_label in [0,1]:
    all_tokens = []
    for row in data:
        if row[0] != class_label:
            continue
        tokens = re.findall(regex_token, row[2])
        all_tokens.extend(tokens)
    print("For class", class_label, ":")
    print("Total number of tokens =", len(all_tokens))
    print("Unique number of tokens =", len(set(all_tokens)))
```

### Question f
count the number of tweets for each day of the week. Eg Mon: 58, Tues: 20,
Wed...

```python
count = [0]*7
```

```python
days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
```

```python
for class_label in [0,1]:
    count = [0]*7
    for row in data:
        if row[0] != class_label:
            continue
        if (re.search(r'Mon', row[1])):
            count[0] += 1
        elif (re.search(r'Tue', row[1])):
            count[1] += 1
        elif (re.search(r'Wed', row[1])):
            count[2] += 1
        elif (re.search(r'Thu', row[1])):
            count[3] += 1
        elif (re.search(r'Fri', row[1])):
            count[4] += 1
        elif (re.search(r'Sat', row[1])):
            count[5] += 1
        elif (re.search(r'Sun', row[1])):
            count[6] += 1
    print('For Class', class_label,'Number of tweets per day:')
    for i in range(7):
        print(days[i] + ": " + str(count[i]))
```

## Part B
### Question a
total number of occurrences of the given word and sentences containing that
word.

```python
class_label = 1
given_word = 'about'
```

```python
regex_word = r'\b{word}\b'.format(word = given_word)
```

```python
regex_sentence = r'\b[^!\?\.]+\b{word}\b[^!\?\.]*[!\?\.]?|\b{word}\b[^!\?\.]*[!\?\.]?'.format(word=given_word)
```

```python
total_occurances = 0
sentences = []
num_sentences = 0
for row in data:
    if row[0] != class_label:
            continue
    occurances = re.findall(regex_word, row[2])
    total_occurances += len(occurances)
    match = re.findall(regex_sentence, row[2])
    sentences.extend(match)
    num_sentences += len(match)
print("For class", class_label, "and given word:", given_word, ":")
print("Number of occurances of word =", total_occurances)
print("Number of sentences =", num_sentences)
print("All sentences are:")
for sent in sentences:
    print(sent)
```

### Question b
number of sentences starting with the given word.

```python
regex_start = r'^\b{word}\b[^!\?\.]*[!\?\.]?'.format(word = given_word)
```

```python
start_sent = []
count = 0
for sent in sentences:
    if re.search(regex_start, sent) is not None:
        start_sent.append(sent)
        count+=1
print("Number of sentences with", given_word, "at the start =", count)
print("The sentences are:")
for s in start_sent:
    print(s)
```

### Question c
number of sentences ending with the given word.

```python
regex_end = r'\b[^!\?\.]*\b{word}\b[!\?\.]|\b[^!\?\.]*\b{word}\b$'.format(word = given_word)
```

```python
end_sent = []
count = 0
for sent in sentences:
    if re.search(regex_end, sent) is not None:
        end_sent.append(sent)
        count+=1
print("Number of sentences with", given_word, "at the end =", count)
print("The sentences are:")
for s in end_sent:
    print(s)
```
